```python
import pandas as pd

# 🔍 Load the HMIS CSV file
hmis_path = r"C:\Users\utkar\OneDrive\Desktop\ClimateXTelemedicine Odisha\Odisha_VScode\.venv\Policy note\Raw\Copy of Monthly Reporting Dashboard Odisha - 2024-25.xlsx - HMIS data (1).csv"
df = pd.read_csv(hmis_path)

# ✅ Basic Overview
print("✅ Total number of records:", len(df))
print("\n🧾 Columns in the dataset:\n", df.columns.tolist())

# 🏥 Number of facilities per district
if 'Districtname' in df.columns:
    district_summary = df['Districtname'].value_counts().sort_index()
    print("\n📍 Facilities per district:\n", district_summary)
else:
    print("⚠️ 'Districtname' column not found!")

# 🏷️ Facility type summary
for col in ['Facilitytype', 'Facilitysubtype']:
    if col in df.columns:
        print(f"\n🏷️ Summary for {col}:\n", df[col].value_counts())
    else:
        print(f"⚠️ '{col}' column not found!")

# 🛑 Duplicates in NIN
if 'NIN' in df.columns:
    duplicates = df[df.duplicated(subset='NIN', keep=False)]
    print(f"\n🛑 Number of duplicate NINs: {duplicates['NIN'].nunique()}")
    print("📋 Sample of duplicate NIN entries:\n", duplicates[['NIN', 'Facilityname', 'Districtname']].head())
else:
    print("⚠️ 'NIN' column not found!")
```

```python
import pandas as pd
import os

# 📂 Input path
input_path = r"C:\Users\utkar\OneDrive\Desktop\ClimateXTelemedicine Odisha\Odisha_VScode\.venv\Policy note\Raw\Copy of Monthly Reporting Dashboard Odisha - 2024-25.xlsx - HMIS data (1).csv"

# 📂 Output path
output_dir = r"C:\Users\utkar\OneDrive\Desktop\ClimateXTelemedicine Odisha\Odisha_VScode\.venv\Policy note\Datasets"
output_filename = "hmis_odisha_2024_cleaned_no_notmapped_nin.csv"
output_path = os.path.join(output_dir, output_filename)

# 📥 Load data
df = pd.read_csv(input_path)

# 🧹 Filter out 'Notmapped' NIN rows
df_cleaned = df[df['NIN'].str.lower() != 'notmapped'].copy()

# ✅ Save the cleaned file
df_cleaned.to_csv(output_path, index=False)
print(f"✅ Cleaned HMIS file saved to:\n📄 {output_path}")
print(f"🧾 Remaining records: {len(df_cleaned)}")
```
